```python
import bq_helper
import pandas as pd
```

```python
session = bq_helper.BigQueryHelper(active_project = 'bigquery-public-data',
                                  dataset_name = 'google_analytics_sample')
```

```python
session.list_tables()
```

```python
session.table_schema('ga_sessions_20170731')
```

```python
session.head('ga_sessions_20170731')
```

```python
query = "SELECT geoNetwork, visitNumber, channelGrouping, socialEngagementType FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170731`"
```

```python
session.estimate_query_size(query)
```

```python
dados = session.query_to_pandas(query)
```

```python
dados.head()
```

```python
geolocation = pd.DataFrame(list(dados.geoNetwork))
geolocation.head()
```

```python
geolocation.info()
```

```python
dados['continent'] = geolocation['continent']
dados['subContinent'] = geolocation['subContinent']
dados['country'] = geolocation['country']
dados['region'] = geolocation['region']
dados['city'] = geolocation['city']
dados['network'] = geolocation['networkDomain']

dados.drop('geoNetwork', axis=1, inplace=True)
```

```python
dados.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados.city.value_counts()
```

```python
dados = dados[dados.city != 'not available in demo dataset']
```

```python
dados = dados[dados.city != '(not set)']
```

```python
dados.city.value_counts()
```

```python
dados.head()
```

```python
dados.network.value_counts()
```

```python
dados = dados[dados.network != '(not set)']
dados = dados[dados.network != 'unknown.unknown']
```

```python
dados_pais = dados.groupby(dados.country).sum()

ax, fig = plt.subplots(figsize=(30,8))
plt.bar(x = dados_pais.index, height = dados_pais.visitNumber)
fig.spines['right'].set_visible(False)
fig.spines['top'].set_visible(False)
plt.title("Visit Number - Country", size=20, color='black', style='oblique', fontweight='bold')
plt.show()
```

```python
dados_regiao = dados.groupby(dados.region).sum()
dados_regiao = dados_regiao[dados_regiao['visitNumber'] > 10]
dados_regiao
```

```python
dados_regiao = dados_regiao[dados_regiao != '(not set)']
```

```python
ax, fig = plt.subplots(figsize=(30,8))
plt.bar(x = dados_regiao.index, height = dados_regiao.visitNumber)
fig.spines['right'].set_visible(False)
fig.spines['top'].set_visible(False)
plt.title("Visit Number - Region > 10", size=20, color='black', style='oblique', fontweight='bold')
plt.show()
```

```python
sub_continent = dados.groupby(dados['subContinent']).sum()
sub_continent
```

```python
ax, fig = plt.subplots(figsize=(30,8))
plt.bar(x = sub_continent.index, height = sub_continent.visitNumber)
fig.spines['right'].set_visible(False)
fig.spines['top'].set_visible(False)
plt.title("Visit Number - Sub Continent", size=20, color='black', style='oblique', fontweight='bold')
plt.show()
```

```python
networks = dados.groupby(dados['network']).sum()
networks = networks[networks['visitNumber'] > 10]
networks
```

```python
ax, fig = plt.subplots(figsize=(30,8))
plt.bar(x = networks.index, height = networks.visitNumber)
fig.spines['right'].set_visible(False)
fig.spines['top'].set_visible(False)
plt.title("Visit Number - Networks", size=20, color='black', style='oblique', fontweight='bold')
plt.show()
```
